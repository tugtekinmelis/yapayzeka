```python
import matplotlib.pyplot as plt#görselleştirme için kullanılır
```

```python
liste=[2,4,1,12,2]
plt.plot(liste)
plt.show()
```

```python
liste=[2,4,1,12,2]
plt.plot(liste,marker='o')
plt.ylabel("Değerler")
plt.xlabel("İndeks")
plt.title("Grafik Tablosu")
plt.show()
```

```python
import numpy as np

x=np.arange(25)
y=x**2+1
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('Çizgi Grafik')
plt.ylabel('Değerler')
plt.xlabel('İndeks')
plt.show()
```

```python
diller=["Python","R","C#","Java"]
oranlar=[60,25,5,10]
plt.xlabel("Diller")
plt.ylabel("Oranlar")
plt.bar(diller,oranlar,color='pink')
plt.show()
```
